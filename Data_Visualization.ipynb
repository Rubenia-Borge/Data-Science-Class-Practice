# Part 1. Simple Charts

# 1. Pie Chart

```
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

#Data to plot
labels = ['Penguins','Polar Bears','Lions','Otters']
sizes = [30,40,20,10]
colors = ['yellow', 'pink', 'purple','red']
explode = [0.1,0.1,0.1,0.1] #explode all slices

#plotting the pie chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors,autopct='%1.1f%%',shadow=True,startangle=270)
plt.title('Distribution of Animals at the Zoo\n',fontweight='bold')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

# 2. Bar Plot

```
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

#Set x-axis labels and data
objects = ('Comedy', 'Action', 'Romance', 'Drama','Sci-Fi')
y_pos = np.arange(len(objects))
performance = [35,15,1,25,40]

#Create bar plot with labels, a title, and color
plt.bar(y_pos, performance, align='center', alpha=0.5,color='pink')
#Places the labels on the x-axis at their various locations determined by np.arange(len(objects))
plt.xticks(y_pos, objects, rotation='70')
plt.ylabel('Number of People')
plt.xlabel('Types of Movies')
plt.title('People who like different types of movies')

#Display the bar plot
plt.show()
```

# 3. Grouped Bar Plots

## 3.1 Side-by-side Bar Plot

```
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

#First set of data
children = (780, 1050, 3056, 5025)
ind = np.arange(4)  # the x locations for the groups
width = 0.35       # the width of the bars
fig, ax = plt.subplots()

#Creating the first set of bars for the first set of data
rects1 = ax.bar(ind, children, width, color='pink') 

#Second set of data-- creation of bars 
adults = (315, 400, 1000, 1500)
rects2 = ax.bar(ind + width, adults, width, color='yellow')

# Add some text for labels, title and axes ticks
ax.set_ylabel('Number of People',fontsize=13)
plt.xlabel('Months',fontsize=12,labelpad=15)
ax.set_title('Number of Visits per Month in the Waterpark',fontsize=13)
ax.set_xticks(ind + width / 2)
ax.set_xticklabels(('April', 'May', 'June', 'July'),fontsize=10)

#Create a legend 
ax.legend((rects1[0], rects2[0]), ('Children', 'Adults'))

plt.show()
```

## 3.2 Stacked Bar Plot

```
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

#Data for the first and second bar plots
children = (780, 1050, 3056, 5025)
adults = (315, 400, 1000, 1500)

#Setting the width of the bars and tick positions
width = 0.5      # the width of the bars
ind = np.arange(1,5)  # the x locations for the groups
tick_pos = [i + (width/50) for i in ind]

#Creating the bars
p1 = plt.bar(ind, children, width, color='pink',align='center')
p2 = plt.bar(ind,adults,width,bottom=children,color='yellow',align='center')

#Add some text for labels, title, axes ticks, and legend
plt.ylabel('Number of People',fontsize=12)
plt.xlabel('Months',fontsize=12,labelpad=15)
plt.title('Number of Visits per Month in the Waterpark',fontsize=13)
plt.xticks(tick_pos,('April','May','June','July'), fontsize=10)
plt.legend((p1[0], p2[0]), ('Children', 'Adults'),loc="best")

#Display the bar plot
plt.show()
```

# Part 3. 3D Plots

# 1. Contour Plot

```
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
#Following line is for color mapping
from matplotlib import cm

#Creates an unfilled contour with a cool & warm color mapping

#Creating the data for the contour plot(s)
delta = 0.025
x = np.arange(0,5,delta)
y = np.arange(0,8,delta)
#meshgrid makes rectangular arrays that
#cover every combination of x and y values 
X, Y = np.meshgrid(x,y)

#make an interesting bump and valley for plotting
#Creates a Gaussian distribution for equal shape X, Y
# Z1 = mlab.bivariate_normal(X,Y,1.0,1.0,0.0,0.0)
# Z2 = mlab.bivariate_normal(X,Y,1.5,0.5,1,1)
Z = np.exp(-((X-1)**2+(Y-4)**2)/0.15) + \
    np.exp(-((X-3)**2+(Y-4)**2)/0.15) + \
    np.exp(-((X-2)**2+(Y-3)**2)/0.15) + \
    np.exp(-(X-2)**2) * np.exp(-(Y - ((X-2)**2+1))**2/0.15)

#Creating an unfilled contour plot
plt.figure()
#Can change the color mappings and line style
cp = plt.contour(X, Y, Z,cmap=cm.coolwarm)

#Labels and titles for the plot
plt.clabel(cp, inline=True, fontsize=10)
plt.title('Contour Plot')
plt.xlabel('X (cm)')
plt.ylabel('Y (cm)')
plt.show()
```

# 2. Mesh Plot

## 2.1 Simple Mesh Plot

```
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np

#Demonstrates plotting a 3D surface colored with the cool-warm color map.
#Surface plot is viewed at a 45-degree angle

#Creating the figure
fig = plt.figure()
ax = fig.gca(projection='3d') #creating axes

#Setting rotation angle of plot
ax.view_init(azim=45) 

# Make data.
X = np.arange(0, 5, 0.25) #array of evenly spaced values
Y = np.arange(0, 8, 0.25) #array of evenly spaced values
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.exp(-((X-1)**2+(Y-4)**2)/0.15) + \
    np.exp(-((X-3)**2+(Y-4)**2)/0.15) + \
    np.exp(-((X-2)**2+(Y-3)**2)/0.15) + \
    np.exp(-(X-2)**2) * np.exp(-(Y - ((X-2)**2+1))**2/0.15)

# Plot the surface.
#The surface is made opaque by using antialiased=False.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,linewidth=0, antialiased=False) 

# Customize the z axis.
ax.set_zlim(-1.01, 1.01)

#Setting the tick locators-- choosing tick locations
#Creates evenly spaced ticks from min to max 
ax.zaxis.set_major_locator(LinearLocator(10)) 
#sprintf format string
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f')) 

# Add a color bar which maps values to colors.
fig.colorbar(surf,shrink=0.6,aspect =10,pad=0.09)

#Add labels for x,y, and z axes
ax.set_xlabel('X axis',fontsize=10)
ax.set_ylabel('Y axis',fontsize=10)
ax.set_zlabel('Z axis',fontsize=10)
ax.set_title('Mesh Plot',fontsize=10)
plt.show()
```
